#Tokenizer
---
Now, the reason why you would need this is that when it comes to training neural networks, you're going to be doing a lot of math and math deals with numbers, and instead of having the words being trained in a neural network, you can actually have the number representing that word and it just makes your life a lot easier.

---
num_words = the tokenizer using the NumWords property or the NumWords 
parameter. In this case, what it's going to do is, in your body of texts that it's tokenizing, it will take the 100 most common words or whatever value that you actually put in here. 

---
oov_token(out of vocab token) = In many cases, it's a good idea to instead of just ignoring unseen words, to put a special value in when an unseen word is encountered.

---
With texts you'll face a similar requirement before you can train with texts, we needed to have some level of uniformity of size, so padding is your friend there.

Default Padding: padding/inserts 0s from left/pre of sentence
padding = 'post': padding 0s from post of sentence

Default max_len: Max length of the string in sentence array
max_len = 5: Change the max len to have 5 words only

So, from where info is lost

Default: pre of sentence
truncating = 'post': lose info from the end

---

```
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

sentences = [
             "i love my dog",
             "I, love my cat",
             "You love my dog!",
             "Do you think my dog is amazing?"
]

tokenizer = Tokenizer(num_words = 100, oov_token='<OOV>')
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index

sequence = tokenizer.texts_to_sequences(sentences)

padded = pad_sequences(sequence, maxlen=5)

print("TRAIN/INITIALIZATION:")
print("\nWord Index = " , word_index)
print("\nSequences = " , sequence)
print("\nPadded Sequences:")
print(padded)

test_data = [
             'I really love my dog',
             'My dog loves my manatee'
]

sequence = tokenizer.texts_to_sequences(test_data)

padded = pad_sequences(sequence, maxlen=10)

print("\nTEST:")
print("\nWord Index = " , word_index)
print("\nSequences = " , sequence)
print("\nPadded Sequences:")
print(padded)
```

#Lesson 2
We'll start with this one published by Rishabh Misra with details on Kaggle at this link. It's a really fun CC0 public domain data-set at all around sarcasm detection. Really? Yeah, really. 

It has three elements in it. The first is sarcastic, is our label. It's a one if the record is considered sarcastic otherwise it's zero. The second is a headline, which is just plain text and the third is the link to the article that the headline describes. 

To make it much easier to load this data into Python, I made a little tweak to the data to look like this.(From indidual set of 3 objects to a list of these objects)

---
So first you need to import JSON. This allows you to load data in JSON format and automatically create a Python data structure from it.

I want the sentences as a list of their own to pass to the tokenizer, I can then create a list of sentences.

```
!wget --no-check-certificate \
    https://storage.googleapis.com/laurencemoroney-blog.appspot.com/sarcasm.json \
    -O /tmp/sarcasm.json
  
import json
with open("/tmp/sarcasm.json", 'r') as f:
    datastore = json.load(f)

headlines = []
labels = []
urls = []

for item in datastore:
  headlines.append(item['headline'])
  labels.append(item['is_sarcastic'])
  urls.append(item['article_link'])
```

If you specify num words to get the top 1000 or whatever, you may be confused by seeing something greater than that here. It's an easy mistake to make, but the key thing to remember, is that when it takes the top 1000 or whatever you specified, it does that in the text to sequence this process.

---
And then finally the shape of the padded one is that each of the sentences in the data set is being padded up to 40 characters long, or 40 words long. And so we have 26,709 sentences in the data set, so my shape of my padded array will be that. And this is what could be used to then train a neural network with embeddings, that you'll be seeing next week.

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

sentences = headlines
tokenizer = Tokenizer(oov_token='<OOV>')
tokenizer.fit_on_texts(sentences)

word_index = tokenizer.word_index
print(len(word_index))
print(word_index)

sequence = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequence, padding='post')
print(sentences[0])
print(padded[0])
print(padded.shape)
```
